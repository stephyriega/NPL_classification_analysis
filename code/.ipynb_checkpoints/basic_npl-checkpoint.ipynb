### Creación de los diccionario estáticos

Como mencionamos en un inicio, las palabras que los cuidadores usen en las respuestas pueden ser similares entre aquellos que responden a una misma respuesta. Este conjunto de palabras usadas en un mismo contexto se le puede llamar "diccionario". Mas aún, ciertas palabras pueden ayudarnos a identificar el tipo de la respuesta (MF/MC/MN). 

En muchos casos, para realizar el proceso de clasificación como el que vamos a hacer, se han utilizado diccionarios que ya tienen establecida una etiqueta para cierto conjunto de palabras (aprendizaje supervisado) o que buscan ordenar estos conjuntos según sus similaridades (aprendizaje no supervisado <a id="3">[3]</a> ). 

En el caso de los primeros, se puede clasificar a ciertos grupos de palabras a partir de relaciones ya establecidas entre una respuesta y su tipo. Esto se ha realizado con información de millones de documentos y que han sido recopilado en sitios web a modo de [open sources](https://huggingface.co/). En otras palabras, se ha usado información en grandes cantidades (normalmente, entre 10M-100M de datos) para establecer que ciertas palabras, y en especial, la relación entre algunas ellas pueden ayudarnos a identificar *sus tipos o etiquetas (qué tan cool es eso, ¿no?)*. Esto, con el proposito de poder clasificar luego a otros conjuntos de palabras nuevas con alguna de las etiquetas ya conocidas en el diccionario. Asimismo, se ha usado igualmente grandes cantidades de información para, sin saber el tipo de , tratar de clasificar estos tipos de respuesta a partir de ver su agrupamiento. Es decir, se buscar *crear etiquetas* a partir del analisis del clustering de los propios conjuntos de palabras.

Sin embargo, para hacer uso de estos diccionarios de aprendizaje supervisado se necesitan que existan y que hayan sido desarrollados en el idioma que usaremos (en nuestro caso, el español). Aún si no existieran, es posible desarrollar nuestros propios diccionarios, pero para ello se necesita información en grandes cantidades. 

Para nuestro ejercicio, no hemos podido encontrar diccionarios ya existentes en español que respondan al contexto de mentalidad del crecimiento <a id="4">[4]</a>  . En cuanto a la posibilidad de crear nuestro propio diccionario, considerarla es más dificil de lo que parece. Primero, porque tenemos un número de pequeño de muestra (~1200). En el caso más optimista, suponiendo que todos nuestros hogares respondan y si tenemos al menos 8 palabras por cada respuesta, tendriamos 9.6M palabras por pregunta (sin considerar que la información verdaderamente útil puede ser mucho menor pues estamos tomando en cuenta los pronombres. Sin embargo, para poder "entrenar el diccionario" necesitariamos usar *solo* una parte de esa muestra, es decir, menos de 9.6M. Tomando en cuenta que los diccionarios colgados en la web usan información de al menos 10K de palabras, no nos parece adecuado seguir esta ruta. 

No obstante, no todo esta perdido. La no existencia actual de un diccionario de este tipo y la posibilidad de no crear uno a partir de estas técnicas no implica que no podamos elaborar uno de manera más simple. En nuestro caso, a partir de una revisión visual de las respuestas <a id="5">[5]</a>  y la clasificación entre ellas, hemos creado diccionarios "estático" (pues no se alimenta con nueva información si es que no la añadimos manualmente) que clasifican una respuesta a un tipo MF/MC/MN dependiendo en el tipo de palabras clave que usa. 

```python
eti = pd.read_excel ('C:\\Users\\Stephy\\Box\\El Salvador growth mindset\\07_Questionnaires&Data\\Endline_Quant\\03_DataManagement\\05 Data analysis\\raw\\Open Questions - Endline\\etiquetas.xlsx', sheet_name="ra_1")
eti.head()

#orden es: las palabras del 0 al 15 son MN, del 16 al 35 son MF y del 36 al 108 son MC. 

#Para ver una palabra en especifico, puede usar
#eti['Lista'][108] 
```

Como podemos ver, hemos generado diccionario estático de palabras para clasificar a las respuestas. Al momento de crearlo, hemos obtenido 250 observaciones, que es aproximadamente un 20% de la muestra, lo cual es positivo pues es una cantidad considerable de todas las respuestas que obtendremos al final de la recolección de datos. Con este diccionario, consideramos que si una respuesta contiene a una de las palabras, se le asignará la etiqueta correspondiente a esa palabra. Por ejemplo, si una respuesta contiene a la palabra "paciencia", esta será clasificada como MN. Para realizar este proceso, creamos una variable "codigo" y le otorgamos un valor (MC=1, MF=2, MN=3) dependiendo del tipo de palabras que contenga.

```python
#creación de la variable
df['codigo'] = df['ra_1']
df.head()
```

```python
#Bucle que recorre cada palabra clave de las etiquetas y codifica

# 3 es para MN, 2 es para MF y 1 es para MC

for j in range(0,15):
  df['codigo'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]),'3', df['codigo'] )
for j in range(16,35):
  df['codigo'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]),'2', df['codigo'] )
#rango de palabras para la categoria 3
for j in range(36,108):
  df['codigo'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]),'1', df['codigo'] )

df.head()
```

Como podemos ver, se ha clasificado las respuestas a partir de los diccionarios establecidos, otorgandoles un valor a cada una dependiendo de su etiqueta. En las líneas de codigo anterior, hemos realizado loop consecutivos, lo cual quiere decir que si primero se identifica una palabra perteneciente a la categoría 3 y luego se identifica una palabra perteneciente a la categoría 2, se quedará con última etiqueta asignada (la de la categoría 2).

No obstante, es importante reconocer nuestras limitaciones. Consideremos que " porque es inteligente y puede aprender más" una posible respuesta a la pregunta 1 del módulo de Percepción de habilidades. Si en nuestro diccionario hemos clasificado a una respuesta que contiene a la respuesta "inteligente" como una del tipo MF, y como una del tipo MC si contiene a la palabra "aprender" *¿Cómo sabemos cual etiqueta elegir?.* Asimismo, tenemos el caso opuesto *¿Qué etiqueta otorgamos cuando ninguna de las palabras esta presente en el diccionario?.*

### Ajustes para la reclasificación de las respuestas abiertas *pues estamos usando un diccionario estático.*

Para afrontar este problema, primero debemos identificar la cantidad de respuestas que lo tienen. Para esto, vamos a crear variables dummies con cada categoría llamadas "codigo_x" y otorgarles el valor de 1 si es que contiene a la palabra perteneciente a su conjunto de palabras. Luego, vamos a identificar a aquellas que tengan un valor distinto de 1 como 0, para aquellas respuestas que ninguna de las palabras pertenezcan al diccionario, o 2, para aquellas respuestas que contengan a dos palabras que pertenecen a distintas etiquetas. 

```python
#creacion de variables dummyes
df['codigo1'] = df['ra_1']
df['codigo2'] = df['ra_1']
df['codigo3'] = df['ra_1']

df.head()
```

```python
for j in range(0,15):
  df['codigo3'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '0' , '0')
for j in range(16,35):
  df['codigo2'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '0' , '0')
#rango de palabras para la categoria 3
for j in range(36,108):
  df['codigo1'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '0' , '0')

for j in range(0,15):
  df['codigo3'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '1' , df['codigo3'] )
for j in range(16,35):
  df['codigo2'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '1' , df['codigo2'] )
#rango de palabras para la categoria 3
for j in range(36,108):
  df['codigo1'] = np.where(df['ra_1'].str.contains(eti['Lista'][j]), '1' , df['codigo1'] )

df.head()
```

```python
#convertimos a float
df['codigo1'] = df['codigo1'].astype(int)
df['codigo2'] = df['codigo2'].astype(int)
df['codigo3'] = df['codigo3'].astype(int)

#sumamos las variables
df['sum'] = df['codigo1'] + df['codigo2']+ df['codigo3']
df.head()
```

Esto puede estar alineado al hecho de que el tipo de mentalidad es fluctuante y aún esta mezclada relativamente en los padres. 

## Tabla de frecuencias de palabras usadas

Para tener un acercamiento más preciso, realizamos una tabla de frecuencias para contabilizar las palabras usadas en las respuestas.

```python
#tokenizamos el objeto de string (identificamos las palabras presentes)
tokens = [token.text for token in doc if not token.is_stop]
print(tokens)
```

```python
#contamos y ordemos de mayor a menor frecuencia
wfreq=Counter(tokens).most_common()
wfreq
```
